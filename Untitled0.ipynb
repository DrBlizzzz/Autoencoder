```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import torch
from torch.utils.data import TensorDataset, DataLoader
from torch.utils.data import Dataset

import pandas as pd
import numpy as np

import ray 
from ray import tune
from ray.tune.schedulers import ASHAScheduler
from ray.tune import Stopper

import seaborn as sns
import matplotlib.pyplot as plt

import os
```

```python
# кастомный класс для обработки датасетов

class DataCustomDataset():

  def __init__(self, path):

    self.dataset = pd.read_csv(path)

    self.lables = self.dataset['Label']
    self.data = self.dataset.drop('Label', axis=1)

    self.size = self.dataset.shape[1]

  # делим в отношении 80/10/10
  # к сожалению в русском сегменте не нашел особо информации в каком отношении делить грамотнее
  # поэтому просто сделал как предлагалось в статье

  def Split(self):

    X_train, X_test, y_train, y_test = train_test_split(self.data, self.lables, test_size=0.2)
    X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5)

    self.TrainData = (X_train, y_train)
    self.ValidationData = (X_valid, y_valid)
    self.TestData = (X_test, y_test)

  # оставляем только нормальные сэмплы

  def FilterAnomaly(self):

    self.NormalTrainData = self.TrainData[0][self.TrainData[1] == 0]
    self.NormalValidationData = self.ValidationData[0][self.ValidationData[1] == 0]

  # заворачиваем датасет в даталоадер
  
  def CreateDataLoader(self, data, labels, batchSize):

    loader = 0

    if len(labels) != 0:

      tensor_X = torch.tensor(data.values)
      tensor_y = torch.tensor(labels.values)
      loader = DataLoader(TensorDataset(tensor_X, tensor_y), batchSize)

    else:

      tensor_X = torch.tensor(data.values)
      loader = DataLoader(TensorDataset(tensor_X), batchSize)

    return loader
```

```python
# непосредственно модель
# оказалось что сигмоида там и правда была лишняя, авторы не думая скопировали модель
# автокодировщика предназначеннную для восстановления изображений

class AutoEncoder(torch.nn.Module):

    def __init__(self, input_dense, latent_size):

        super().__init__()

        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(input_dense, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, 32),
            torch.nn.ReLU(),
            torch.nn.Linear(32, latent_size),
            torch.nn.ReLU()
        )

        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(latent_size, 32),
            torch.nn.ReLU(),
            torch.nn.Linear(32, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, input_dense),
        )

    def forward(self, x):

        encoded = self.encoder(x)
        decoded = self.decoder(encoded)

        return decoded
```

```python
# размер батча решил оставить статичным, он не так сильно влияет на обучение

batch_size = 512

dataset_creator = DataCustomDataset(path='/content/drive/MyDrive/Colab Notebooks/NSL_KDD_checked.csv')

dataset_creator.Split()
dataset_creator.FilterAnomaly()

normal_train_loader_batched = dataset_creator.CreateDataLoader(data = dataset_creator.NormalTrainData, 
                                                       labels = list(), 
                                                       batchSize = batch_size)

validation_loader_batched = dataset_creator.CreateDataLoader(data = dataset_creator.ValidationData[0], 
                                                     labels = dataset_creator.ValidationData[1],
                                                     batchSize = batch_size)


validation_loader_unbatched = dataset_creator.CreateDataLoader(data = dataset_creator.ValidationData[0], 
                                                     labels = dataset_creator.ValidationData[1],
                                                     batchSize = 1)

normal_validation_loader_unbatched = dataset_creator.CreateDataLoader(data = dataset_creator.NormalValidationData, 
                                                            labels = list(),
                                                            batchSize = 1)

test_loader_unbatched = dataset_creator.CreateDataLoader(data = dataset_creator.TestData[0], 
                                               labels = dataset_creator.TestData[1],
                                               batchSize = 1)
```

```python
# это класс с условием на early-stopping, он используется в подборе гиперпараметров 
# если ошибка больше 5 предыдущих, то заканчиваем обучение

class EarlyStopper(Stopper):

  def __init__(self):
      
    self.sessions = dict()

  def __call__(self, trial_id, result):

    if trial_id in self.sessions:

      if len(self.sessions[trial_id]) < 5:

        self.sessions[trial_id].append(result['loss'])

        return False

      else:

        if len([loss for loss in self.sessions[trial_id] if loss < result['loss']]) == 5:

          return True

        else:

          self.sessions[trial_id].pop(0)

          self.sessions[trial_id].append(result['loss'])

          return False

    else:

      self.sessions[trial_id] = list()

      return False

  def stop_all(self):

    return False
```

```python
# количество эпох также статично(обычно 100 эпох с головой хватает)
# если нужно меньше, то останов обеспечит early-stopping
# таким образом нет необходимости подбирать этот гиперпараметр

epochs = 100

device = 'cuda' if torch.cuda.is_available() else 'cpu'

loss_function = torch.nn.MSELoss()

# класс для проверки качества на трейне
# обучение происходит в два этапа
# первый: обучиться только на нормальных образцах из трейна
# второй: проверить качество на валидации не пересчитывая градиенты

class AutoEncoderTrainable(tune.Trainable):

  def setup(self, config, train_loader, valid_loader, input_dense, device, loss_function):

    self.train_loader = train_loader
    self.valid_loader = valid_loader
    self.input_dense = input_dense

    self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
    self.model = AutoEncoder(input_dense, config['latent_size']).to(self.device)
    self.loss_function = torch.nn.MSELoss()
    self.optimizer = torch.optim.Adam(self.model.parameters(), 
                                      lr=config['lr'], 
                                      weight_decay=config['weight_decay'])

  def step(self):

    for sample in self.train_loader:

      self.optimizer.zero_grad()

      cuda_sample = sample[0].to(self.device).float()

      reconstructed = torch.FloatTensor().to(self.device)
      reconstructed = self.model(cuda_sample)

      loss = self.loss_function(reconstructed, cuda_sample)

      loss.backward()

      self.optimizer.step()

    sum_loss = 0

    for sample in self.valid_loader:

      with torch.no_grad():

        cuda_sample0 = sample[0].to(self.device).float()

        reconstructed = torch.FloatTensor().to(self.device)
        reconstructed = self.model(cuda_sample0)

        loss = self.loss_function(reconstructed, cuda_sample0)

        sum_loss += loss.item()

    return {'loss': sum_loss}

  def save_checkpoint(self, tmp_checkpoint_dir):

      checkpoint_path = os.path.join(tmp_checkpoint_dir, "model.pth")

      torch.save(self.model.state_dict(), checkpoint_path)

      return tmp_checkpoint_dir

# задали пространство поиска гиперпараметров

cfg = {
    'lr': tune.loguniform(1e-5, 1e-2),
    'weight_decay': tune.loguniform(1e-5, 1e-2),
    'latent_size': tune.choice([i + 1 for i in range(32)])
}

# подбор осуществляю с помошью асинхронного половинного деления
# почитал про него, идея понравилась, работает лучше случайного поиска или перебора по сетке

scheduler = ASHAScheduler(
    metric="loss",
    mode="min",
    reduction_factor=2,
    max_t=epochs
    )

analysis = tune.run(
    tune.with_parameters(AutoEncoderTrainable, 
                         train_loader = normal_train_loader_batched, 
                         valid_loader = validation_loader_batched,
                         input_dense = (dataset_creator.size - 1),
                         device = device,
                         loss_function = loss_function),
    resources_per_trial={'gpu': 0.5},
    config=cfg,
    scheduler=scheduler,
    num_samples=64,
    checkpoint_freq=5,
    local_dir='/content/drive/MyDrive/Colab Notebooks/models',
    stop=EarlyStopper()
)


best_trial = analysis.get_best_trial(metric='loss', mode='min')

best_checkpoint = analysis.get_best_checkpoint(best_trial, metric='loss', mode='min')
best_checkpoint = os.path.join(best_checkpoint, 'model.pth')

best_parameters = analysis.get_best_config(metric='loss', mode='min') 

# здесь просто инициализируем лучшими параметрами модель и оптимизатор

model = AutoEncoder(dataset_creator.size - 1, best_parameters['latent_size'])
model.load_state_dict(torch.load(best_checkpoint))
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), 
                             lr=best_parameters['lr'], 
                             weight_decay=best_parameters['weight_decay'])

```

```python
# далее необходимо найти мат ожидание и дисперсию нормальных сэмплов на валидации
# тк мы не обучаемся градиенты не считаем

normal_validation_losses = np.array([])

for sample in normal_validation_loader_unbatched:

  with torch.no_grad():

    cuda_sample0 = sample[0].to(device).float()

    reconstructed = torch.FloatTensor().to(device)
    reconstructed = model(cuda_sample0)

    loss = loss_function(reconstructed, cuda_sample0)

    normal_validation_losses = np.append(normal_validation_losses, loss.item())

mean = normal_validation_losses.mean()
variance = normal_validation_losses.var()

print(mean, variance)

# в этой части рассчитываем z-оценки для каждого обьекта из валидации
# z-оценка показывает в скольких стандартных отклонениях находится отмасштабированный обьект от мат ож

validation_losses = np.array([])

for sample in validation_loader_unbatched:

  with torch.no_grad():

    cuda_sample0 = sample[0].to(device).float()

    reconstructed = torch.FloatTensor().to(device)
    reconstructed = model(cuda_sample0)

    loss = loss_function(reconstructed, cuda_sample0)
  
    validation_losses = np.append(validation_losses, (loss.item() - mean) / (variance ** (0.5) ) )

# на основании полученных z-оценок ищем границу разделения классов

scores = list()

sorted_dimension = list(zip(validation_losses, dataset_creator.ValidationData[1]))
sorted_dimension.sort()

for threshold in np.linspace(-4, 4, 800):

  answers = list()
  true_answers = list()

  for pair in sorted_dimension:

    if pair[0] < threshold:

      answers.append(0)

      true_answers.append(pair[1])

    else:

      answers.append(1)

      true_answers.append(pair[1])
  
  tn, fp, fn, tp = confusion_matrix(true_answers, answers).ravel()

  positive_precision = tp / (tp + fp)
  positive_recall = tp / (tp + fn)
  positive_f1_score = (positive_precision * positive_recall) / (positive_precision + positive_recall)

  negative_precision = tn / (tn + fn)
  negative_recall = tn / (tn + fp)
  negative_f1_score = (negative_precision * negative_recall) / (negative_precision + negative_recall)

  scores.append( (threshold, positive_f1_score, negative_f1_score) )

print(scores)

# максимизируя сумму f1-score по каждому из класссов подбираем THRESHOLD
# THRESHOLD - точка разбиения наших классов

max_point = (0, 0, 0)

for score in scores:

  if sum(score[1:]) > sum(max_point[1:]):

    max_point = score

THRESHOLD = max_point[0]

results = list()

# здесь проверка качества

for test_sample in test_loader_unbatched:

  with torch.no_grad():

    cuda_sample0 = test_sample[0].to(device).float()

    reconstructed = torch.FloatTensor().to(device)
    reconstructed = model(cuda_sample0)

    loss = loss_function(reconstructed, cuda_sample0)

    if ((loss.item() - mean) / (variance ** (0.5))) <  THRESHOLD:

      results.append(0)

    else:

      results.append(1)

print(classification_report(list(dataset_creator.TestData[1]), results))

tn, fp, fn, tp = confusion_matrix(list(dataset_creator.TestData[1]), results).ravel()

print('Ошибок 1-ого рода:', fp)
print('Ошибок 2-ого рода:', fn)



```
